```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
vertices = range(1,10)
edges = [(7, 2), (2, 3), (7, 4), (4, 5), (7, 3), (7, 5), (1, 6), (1, 7), (2, 8), (2, 9)]

## data Latihan 1
vertices1 = range(1,10)
edges1 = [(1, 7), (1, 2), (2, 3), (2, 6), (3, 4), (3, 6), (5, 6), (6, 7), (7, 8), (7, 9), (7, 10), (8, 9)]

## data Latihan 2
vertices2 = range(1,13)
edges2 = [(1, 2), (1, 3), (2, 3), (2, 12), (3, 4), (3, 10), (3, 11), (3, 12), 
            (4, 5), (4, 6), (4, 7), (5, 6), (5, 7), (5, 8), (6, 9), (7, 8), (8, 9), (10, 11), (11, 12), (12, 13)]
```

```python
## Kasus Fraud Analytics
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)
pos = nx.spring_layout(G)

## graph Latihan 1
G1 = nx.Graph()
G1.add_nodes_from(vertices1)
G1.add_edges_from(edges1)
pos1 = nx.spring_layout(G1)

## graph Latihan 2
G2 = nx.Graph()
G2.add_nodes_from(vertices2)
G2.add_edges_from(edges2)
pos2 = nx.spring_layout(G2)
```

```python
nx.draw_networkx_nodes(G, pos, nodelist = [1, 4, 3, 8, 9], node_color = 'g', node_size = 1300)
nx.draw_networkx_nodes(G, pos, nodelist = [2, 5, 6, 7], node_color = 'r', node_size = 1300)
nx.draw_networkx_edges(G, pos, edges, width = 3, alpha = 0.5, edge_color = 'b')
labels = {}
labels[1] = r'1 NF'
labels[2] = r'2 F'
labels[3] = r'3 NF'
labels[4] = r'4 NF'
labels[5] = r'5 F'
labels[6] = r'6 F'
labels[7] = r'7 F'
labels[8] = r'8 NF'
labels[9] = r'9 NF'

nx.draw_networkx_labels(G, pos, labels, font_size = 16)
```
